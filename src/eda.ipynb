## Feature Engineering

```python
# Merge affordability and college results on institution name
merged = college_results_data.merge(
    affordability_gap_data,
    left_on='Institution Name',
    right_on='Institution Name',
    how='left'
)

# 1. Real Net Price (latest available)
merged['real_net_price'] = merged['Average Net Price After Grants, 2020-21']

# 2. Tuition-to-Income Ratio
merged['tuition_to_income_ratio'] = merged['Average In-State Tuition for First-Time, Full-Time Undergraduates'] / (merged['State Minimum Wage']*2080)  # assume 40hrs/week full year

# 3. Parent Penalty (extra cost for student-parent)
merged['parent_penalty_center'] = merged['Student Parent Affordability Gap: Center-Based Care'] - merged['100% TTD Affordability Gap']
merged['parent_penalty_home'] = merged['Student Parent Affordability Gap: Home-Based Care'] - merged['100% TTD Affordability Gap']

# 4. Equity in STEM (example for Black and Latino students)
merged['stem_equity_black'] = merged['Percent of Bachelor Degrees Awarded in Science, Technology, Engineering, and Math'] - merged['Percent of Bachelor Degrees Black or African American Total']
merged['stem_equity_latino'] = merged['Percent of Bachelor Degrees Awarded in Science, Technology, Engineering, and Math'] - merged['Percent of Bachelor Degrees Latino Total']

# 5. Completion Gap Index
merged['completion_gap_black'] = merged["Bachelor's Degree Graduation Rate Within 6 Years - White Non-Latino"] - merged["Bachelor's Degree Graduation Rate Within 6 Years - Black, Non-Latino"]
merged['completion_gap_latino'] = merged["Bachelor's Degree Graduation Rate Within 6 Years - White Non-Latino"] - merged["Bachelor's Degree Graduation Rate Within 6 Years - Latino"]

# 6. Mobility Return Index
merged['mobility_return'] = (
    merged['Median Earnings of Students Working and Not Enrolled 10 Years After Entry'] 
    - merged['Median Debt for Dependent Students'] 
    - merged['real_net_price']
)

# 7. Admissions Competitiveness
merged['sat_score_avg'] = (merged['SAT Evidence Based Reading and Writing - 25th Percentile Score'] + merged['SAT Math - 25th Percentile Score']) / 2
merged['act_score_avg'] = (merged['ACT English - 25th Percentile Score'] + merged['ACT Math - 25th Percentile Score']) / 2
merged['admissions_competitiveness'] = (
    ((merged['sat_score_avg'] - merged['sat_score_avg'].min()) / (merged['sat_score_avg'].max() - merged['sat_score_avg'].min())) +
    ((merged['act_score_avg'] - merged['act_score_avg'].min()) / (merged['act_score_avg'].max() - merged['act_score_avg'].min())) +
    (1 - merged['Total Percent of Applicants Admitted']/100)
) / 3

# 8. Instructional Spending Efficiency
merged['spending_efficiency'] = merged['Instructional Expenses Per FTE'] / merged["Bachelor's Degree Graduation Rate Bachelor Degree Within 6 Years - Total"]

# Preview some key engineered features
merged[['Institution Name', 'real_net_price', 'tuition_to_income_ratio', 
              'parent_penalty_center', 'parent_penalty_home', 'completion_gap_black', 
              'mobility_return', 'admissions_competitiveness', 'spending_efficiency']].head()
```

## Clustering

```python
from sklearn.preprocessing import StandardScaler

feature_cols = [
    'real_net_price', 'tuition_to_income_ratio', 'parent_penalty_center', 'parent_penalty_home',
    'stem_equity_black','stem_equity_latino','completion_gap_black','completion_gap_latino',
    'mobility_return','admissions_competitiveness','spending_efficiency'
]

X = merged[feature_cols].copy()

# Replace inf/-inf with NaN
X.replace([np.inf, -np.inf], np.nan, inplace=True)

# Fill missing values (NaN) with median of each column
X = X.fillna(X.median())

# Standardize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], alpha=0.6)
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("PCA: Colleges across affordability, equity, mobility")
plt.show()
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_tsne = tsne.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
plt.scatter(X_tsne[:,0], X_tsne[:,1], alpha=0.6)
plt.xlabel("t-SNE 1")
plt.ylabel("t-SNE 2")
plt.title("t-SNE: Non-linear embedding of colleges")
plt.show()
```

```python
import umap

reducer = umap.UMAP(
    n_neighbors=15,       # typical range: 10-50, controls local vs global structure
    min_dist=0.1,         # typical range: 0.0-0.5, tighter clustering with smaller
    n_components=2,
    metric='euclidean',
    random_state=42
)
X_umap = reducer.fit_transform(X_scaled)

# Plot
plt.figure(figsize=(10,7))
sns.scatterplot(
    x=X_umap[:,0],
    y=X_umap[:,1],
    alpha=0.7
)
plt.title("UMAP embedding of colleges")
plt.xlabel("UMAP 1")
plt.ylabel("UMAP 2")
plt.show()
```

```python
from sklearn.cluster import KMeans

n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

merged['cluster'] = clusters

# Plot clusters on UMAP
plt.figure(figsize=(8,6))
sns.scatterplot(x=X_umap[:,0], y=X_umap[:,1], hue=merged['cluster'], palette='tab10', alpha=0.7)
plt.title("UMAP + KMeans Clusters of Colleges")
plt.xlabel("UMAP 1")
plt.ylabel("UMAP 2")
plt.legend(title="Cluster")
plt.show()
```
